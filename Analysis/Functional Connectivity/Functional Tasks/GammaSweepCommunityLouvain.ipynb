```python
#choose sample networks
#1 - most sparse
nw1=NWN[0][0]
#2 - middle
nw2=NWN[16][0]
#3 - most dense
nw3=NWN[-1][0]

#some more networks of each density:
nw1a=NWN[0][5]
nw2a=NWN[16][5]
nw3a=NWN[-1][5]

#some more networks of each density:
nw1b=NWN[0][-1]
nw2b=NWN[16][-1]
nw3b=NWN[-1][-1]

avgDegs=[np.array([i for i in nw1['Graph Theory']['Degree']]).T[1],np.array([i for i in nw2['Graph Theory']['Degree']]).T[1],np.array([i for i in nw3['Graph Theory']['Degree']]).T[1]]
```

```python
c1i=[];c2i=[];c3i=[]
c1ai=[];c2ai=[];c3ai=[]
c1bi=[];c2bi=[];c3bi=[]
for i in tqdm(range(100)):
    for gamma in np.arange(0.5,2.1,0.1):
        c1i.append(bct.community_louvain(nw1['adj_matrix'],gamma=gamma))
        c2i.append(bct.community_louvain(nw2['adj_matrix'],gamma=gamma))
        c3i.append(bct.community_louvain(nw3['adj_matrix'],gamma=gamma))
        c1ai.append(bct.community_louvain(nw1a['adj_matrix'],gamma=gamma))
        c2ai.append(bct.community_louvain(nw2a['adj_matrix'],gamma=gamma))
        c3ai.append(bct.community_louvain(nw3a['adj_matrix'],gamma=gamma))
        c1bi.append(bct.community_louvain(nw1b['adj_matrix'],gamma=gamma))
        c2bi.append(bct.community_louvain(nw2b['adj_matrix'],gamma=gamma))
        c3bi.append(bct.community_louvain(nw3b['adj_matrix'],gamma=gamma))
```

```python
ci1=[];q1=[]
ci2=[];q2=[]
ci3=[];q3=[]
ci1a=[];q1a=[]
ci2a=[];q2a=[]
ci3a=[];q3a=[]
ci1b=[];q1b=[]
ci2b=[];q2b=[]
ci3b=[];q3b=[]
for i in range(len(c1i)):
    ci1.append(c1i[i][0])
    q1.append(c1i[i][1])
    ci2.append(c2i[i][0])
    q2.append(c2i[i][1])
    ci3.append(c3i[i][0])
    q3.append(c3i[i][1]) 
    ci1a.append(c1ai[i][0])
    q1a.append(c1ai[i][1])
    ci2a.append(c2ai[i][0])
    q2a.append(c2ai[i][1])
    ci3a.append(c3ai[i][0])
    q3a.append(c3ai[i][1]) 
    ci1b.append(c1bi[i][0])
    q1b.append(c1bi[i][1])
    ci2b.append(c2bi[i][0])
    q2b.append(c2bi[i][1])
    ci3b.append(c3bi[i][0])
    q3b.append(c3bi[i][1]) 
    
    
ci1=np.array(ci1).reshape(100,16,nw1['number_of_wires'][0][0])
q1=np.array(q1).reshape(100,16)
ci2=np.array(ci2).reshape(100,16,nw2['number_of_wires'][0][0])
q2=np.array(q2).reshape(100,16)
ci3=np.array(ci3).reshape(100,16,nw3['number_of_wires'][0][0])
q3=np.array(q3).reshape(100,16)

ci1a=np.array(ci1a).reshape(100,16,nw1a['number_of_wires'][0][0])
q1a=np.array(q1a).reshape(100,16)
ci2a=np.array(ci2a).reshape(100,16,nw2a['number_of_wires'][0][0])
q2a=np.array(q2a).reshape(100,16)
ci3a=np.array(ci3a).reshape(100,16,nw3a['number_of_wires'][0][0])
q3a=np.array(q3a).reshape(100,16)


ci1b=np.array(ci1b).reshape(100,16,nw1b['number_of_wires'][0][0])
q1b=np.array(q1b).reshape(100,16)
ci2b=np.array(ci2b).reshape(100,16,nw2b['number_of_wires'][0][0])
q2b=np.array(q2b).reshape(100,16)
ci3b=np.array(ci3b).reshape(100,16,nw3b['number_of_wires'][0][0])
q3b=np.array(q3b).reshape(100,16)
```

```python
pd1=[[] for i in range(16)];pd2=[[] for i in range(16)];pd3=[[] for i in range(16)]
pd1a=[[] for i in range(16)];pd2a=[[] for i in range(16)];pd3a=[[] for i in range(16)]
pd1b=[[] for i in range(16)];pd2b=[[] for i in range(16)];pd3b=[[] for i in range(16)]

for k in tqdm(range(len(np.arange(0.5,2.1,0.1)))):
    for i in range(100):
        for j in range(100):
            pd1[k].append(bct.partition_distance(ci1[i][k],ci1[j][k]))
            pd2[k].append(bct.partition_distance(ci2[i][k],ci2[j][k]))
            pd3[k].append(bct.partition_distance(ci3[i][k],ci3[j][k]))
            pd1a[k].append(bct.partition_distance(ci1a[i][k],ci1a[j][k]))
            pd2a[k].append(bct.partition_distance(ci2a[i][k],ci2a[j][k]))
            pd3a[k].append(bct.partition_distance(ci3a[i][k],ci3a[j][k]))
            pd1b[k].append(bct.partition_distance(ci1b[i][k],ci1b[j][k]))
            pd2b[k].append(bct.partition_distance(ci2b[i][k],ci2b[j][k]))
            pd3b[k].append(bct.partition_distance(ci3b[i][k],ci3b[j][k]))
```

```python
MI1=[];MI2=[];MI3=[]
MI1a=[];MI2a=[];MI3a=[]
MI1b=[];MI2b=[];MI3b=[]

for i in range(len(pd1)):
    for j in range(len(pd1[i])):
        MI1.append(pd1[i][j][1])
        MI2.append(pd2[i][j][1])
        MI3.append(pd3[i][j][1])
        MI1a.append(pd1a[i][j][1])
        MI2a.append(pd2a[i][j][1])
        MI3a.append(pd3a[i][j][1])
        MI1b.append(pd1b[i][j][1])
        MI2b.append(pd2b[i][j][1])
        MI3b.append(pd3b[i][j][1])
        
MI1=np.array(MI1).reshape(16,10000)
MI2=np.array(MI2).reshape(16,10000)
MI3=np.array(MI3).reshape(16,10000)    
MI1a=np.array(MI1a).reshape(16,10000)
MI2a=np.array(MI2a).reshape(16,10000)
MI3a=np.array(MI3a).reshape(16,10000)   
MI1b=np.array(MI1b).reshape(16,10000)
MI2b=np.array(MI2b).reshape(16,10000)
MI3b=np.array(MI3b).reshape(16,10000)   
```

```python
x=gammas
plt.plot(x,np.nanmean([np.nanmean(MI1,axis=1),np.nanmean(MI1a,axis=1),np.nanmean(MI1b,axis=1)],axis=0),'-o')
y=np.nanmean([np.nanmean(MI1,axis=1),np.nanmean(MI1a,axis=1),np.nanmean(MI1b,axis=1)],axis=0)
err=np.nanstd([np.nanmean(MI1,axis=1),np.nanmean(MI1a,axis=1),np.nanmean(MI1b,axis=1)],axis=0)
plt.fill_between(x,y-err,y+err,alpha=0.3)

plt.plot(x,np.nanmean([np.nanmean(MI2,axis=1),np.nanmean(MI2a,axis=1),np.nanmean(MI2b,axis=1)],axis=0),'-o')
y=np.nanmean([np.nanmean(MI2,axis=1),np.nanmean(MI2a,axis=1),np.nanmean(MI2b,axis=1)],axis=0)
err=np.nanstd([np.nanmean(MI2,axis=1),np.nanmean(MI2a,axis=1),np.nanmean(MI2b,axis=1)],axis=0)
plt.fill_between(x,y-err,y+err,alpha=0.3)

plt.plot(x,np.nanmean([np.nanmean(MI3,axis=1),np.nanmean(MI3a,axis=1),np.nanmean(MI3b,axis=1)],axis=0),'-o')
y=np.nanmean([np.nanmean(MI3,axis=1),np.nanmean(MI3a,axis=1),np.nanmean(MI3b,axis=1)],axis=0)
err=np.nanstd([np.nanmean(MI3,axis=1),np.nanmean(MI3a,axis=1),np.nanmean(MI3b,axis=1)],axis=0)
plt.fill_between(x,y-err,y+err,alpha=0.3)

plt.xlabel(r'$\gamma$')
plt.ylabel('Mutual Information')
h=plt.legend([str(np.round(np.mean(avgDegs[0]))),str(np.round(np.mean(avgDegs[1]))),str(np.round(np.mean(avgDegs[2])))])
h.set_title('Avg Degree',prop={'weight':'bold'})

saveLoc='/import/silo2/aloe8475/Documents/CODE/Data/Figures/Functional Connectivity/'

plt.savefig(saveLoc+'GammaSweepCommunityLouvainDifferentDensities - Mutual Information vs Gamma.pdf',format='pdf',dpi=300)
```

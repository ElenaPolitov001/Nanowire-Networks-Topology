```python
def nonlineartrans(network,i,onAmp,nodesList,stimulus):
    
    dt = 1e-2
    f=0.5
    Time=5
    period=1/f
   #Instantiate Variables    
    print('Running Linear Transformation Simulations')
    dt = 1e-2
    f=0.5
    Time=5

    #Choose Electrode Pattern
    stimulus[0].append((stimulus__(biasType='AC',onAmp=onAmp,TimeVector=np.arange(0,Time,dt),f=f)))
    stimulus[1].append((stimulus__(biasType='Drain',TimeVector=np.arange(0,Time,dt)))) #we don't want this drain to be active during training

    #Initialise Output Variables
    period=[]
    TimeVector=[]
    voltage=[]
    conductance=[]
    switches=[]
    results=[]
    #Run Simulations
#     results=[]
    # Connectivity=connectivity__('700nw_14533junctions.mat')
    stimulus2 = [item for item in stimulus] #go through each list in the list and find the ith item
#     set_trace()
    results=runSim(connectivity__(graph=network['G']),stimulus=stimulus2, contactMode='farthest', T = Time, dt = 0.001, onAmp = onAmp, biasType='AC',f=f,junctionMode='tunneling')
    #wires_dict=newNetworkTest[chosenNetwork])
    results.frequency=f
    results.dt=0.001
#     print('Saving 10 Simulations for Network ' +str(i+1))
#     resultsNew={'JunctionConductance':results.junctionConductance[-1],'JunctionVoltage':results.junctionVoltage[-1],'WireVoltage':results.wireVoltage[-1],'Switch':results.junctionSwitch[-1],'Source':results.sources[0],'Drain':results.drains[0]}
#     compressed_pickle(resultsNew,'/import/silo2/aloe8475/Documents/CODE/Data/Functional Connectivity/Hierarchical Results/Hierarchical_NLT_Simulation_2VAmp_'+str(i+1)+'.pkl')
#     print('Saved')
    period=1/f  

    TimeVector=results.TimeVector
    voltage=results.wireVoltage
    conductance=results.conductance
    switches=results.junctionSwitch

    target1= (onAmp * (-np.sign(TimeVector % period - period/2)))
#             target2[j] = (4*onAmp[i]/period * abs((TimeVector-period/4) % period - period/2) - onAmp[i])
#             target3[j] = (onAmp[i]/period * (TimeVector % period))
#             target4[j] = (onAmp[i]*np.sin(4*np.pi*(1/period)*TimeVector))

    if len(network['G']) >= 250:
        nodesList=[50,100,150,200,250,len(network['G'])]#range(50, len(ws300[i][0])+1,50)
    elif len(network['G']) >= 200 and len(network['G']) < 250:
        nodesList=[50,100,150,200,len(network['G'])]#range(50, len(ws300[i][0])+1,50)  
    elif len(network['G']) >= 150 and len(network['G']) < 200:
        nodesList=[50,100,150,len(network['G'])]#range(50, len(ws300[i][0])+1,50)  
    elif len(network['G']) >= 100 and len(network['G']) < 150:
        nodesList=[50,100,len(network['G'])]#range(50, len(ws300[i][0])+1,50)
    
    nwSqu =[None]*len(nodesList)
    countK=0

    for k in nodesList: #loop through sets of nodes for regression
        print('Running Regression: ' + str(k) + ' nodes')
        ResultSqu=[]
        ResultSqu = nonLinearTrans(results,'Square',k, repeats=50) #simulation, type of signal, number of nodes to sample from, number of linear regression repetitions (take avg)
        nwSqu[countK]=ResultSqu['accuracy']
#         network['Accuracy']['NLT'][countK]=nwSqu[countK]
        countK=countK+1
            #Save networks so we don't have to run this every time
        
    return nwSqu
```

```python
onAmp=[]
shortestPath=[]
adjMats={"AdjMat":[None]*len(newHNW)}

for i in tqdm(range(len(newHNW))):
    ci=[]
    q=[]
    
    temp=getFarthestPairing(newHNW[i]['AdjMat'])
    adjMats['AdjMat'][i]=newHNW[i]['AdjMat']
    
    shortestPath=nx.shortest_path_length(newHNW[i]['G'],temp[0],temp[1])
    onAmp.append(shortestPath/5) #need to justify divide by 5 or come up with some other way
# HNW['sp_amp']=onAmp
    newHNW[i]['Graph Theory']['Shortest Path']=shortestPath
    ci,q=nct.community_louvain(newHNW[i]['AdjMat'])
    newHNW[i]['Graph Theory']['Modularity Score']=q
    newHNW[i]['Graph Theory']['AvgDeg']=(np.mean(np.asarray(newHNW[i]['G'].degree())[:,1]))
    newHNW[i]['Graph Theory']['PCoeff']=bct.participation_coef(newHNW[i]['AdjMat'],ci)
    newHNW[i]['Graph Theory']['MZ']=bct.module_degree_zscore(newHNW[i]['AdjMat'],ci)

savemat('HierarchicalNW_AdjMat.mat',adjMats)
```

```python
def get_MC_new(Connectivity,i, amp = 2, dt = 0.01, dur = 1, n_outs = 100, init_state = None, return_sim = False):
    signal = signal_expand(np.random.random(1000)*2*amp-amp, dur)
    simRAND = runSim(Connectivity, T = 10, dt = dt,
                start_state =  init_state,
                biasType = 'Custom', customSignal = signal, 
                junctionMode = 'tunneling', 
                contactMode = 'farthest',
                findFirst=False, disable_tqdm=True)
#     print('Saving Simulations')
#     results={'JunctionConductance':simRAND.junctionConductance,'JunctionVoltage':simRAND.junctionVoltage,'Switch':simRAND.junctionSwitch,'conductance':simRAND.conductance,'Source':simRAND.sources[0],'Drain':simRAND.drains[0],'Signal':simRAND.stimulus}
#     compressed_pickle(results,'/import/silo2/aloe8475/Documents/CODE/Data/Functional Connectivity/Hierarchical Networks/Hierarchical_MC_Simulation_2VAmp_'+str(i+1)+'.pkl')
    mcRAND = [MC_k(simRAND, i, n_outs = n_outs) for i in range(dur, 101*dur, dur)]

    if return_sim:
        return np.sum(mcRAND), simRAND
    else:
        return np.sum(mcRAND)
```

```python
init=[]
for i in range(len(newHNW)):
    amp=2#onAmp[i][j]
    connectivity=connectivity__(graph=newHNW[i]['G'])
    init.append(inputPacker(get_MC_new,connectivity,i,amp))
```

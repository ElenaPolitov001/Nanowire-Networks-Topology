```python
def genGridNW(xa,xb,ya,yb,ex,ey):
    e = []
    for i in range(len(ex)):
        d = np.zeros(len(xa))
        for j in range(len(xa)):
            d[j]=dist((xa[j], ya[j]), (xb[j], yb[j]), (ex[i], ey[i]))
        e.append(np.argmin(d))
    return np.array(e)

def point_on_line(a, b, p):
    ap = p - a
    ab = b - a
    result = a + np.dot(ap, ab) / np.dot(ab, ab) * ab
    return result

def dist(p1,p2,p3):
    p1=np.array(p1) #xa ya
    p2=np.array(p2) #xb yb
    p3=np.array(p3) # ex ey (electrode placement)
    
    #determine whether closest point to electrode is on the line, or outside the line 
    t=point_on_line(p1,p2,p3)
    if t[1]<p1[1]:
        r = np.linalg.norm(p3-p1) #if point is outside left (xvalues)
    elif t[1] > p2[1]:
        r = np.linalg.norm(p3-p2) #if point is outside right (xvalues)
    else:
        r = np.linalg.norm(np.cross(p2-p1, p1-p3))/np.linalg.norm(p2-p1) #if point is inside xvalues
    return r
```

```python
def init(sources,onAmp_init,f):
    #Initialise stimulus:
    
    init_stimulus = []

    init_sims=[None]*numinit


    #init STIMULUS
    # ONLY DRAINS WITH CORRESPONDING STIMULUS SOURCE ARE ACTIVE
    #Drains: 
    for i in range(num_drain_init):
        #DRAIN WITHOUT REST
        init_stimulus.append((stimulus__(biasType='Drain',T=run_time,dt=dt)))

    #Sources
    for i in range(num_source_init):
        #SOURCE WITHOUT REST
        init_stimulus.append((stimulus__(biasType='Square',f=f,onAmp=onAmp_init,T=run_time,dt=dt)))
    
    return init_stimulus
```

```python
#Get the state of the network when just N drains are open, where N = number of target drains
def getInitState(init_stimulus,sources,drain_pool,target_index):
    eles = np.append(drain_pool[target_index], sources)
    if len(eles) == num_drain_init + num_source_init:
        init_sim = runSim(connectivity, stimulus = init_stimulus,
                           junctionMode = 'tunneling',
                           dt = dt, T = run_time, 
                           contactMode = 'preSet',
                           electrodes = eles,
                           findFirst = False,disable_tqdm=True,
                           criticalFlux=0.01)

        JS0 = getJunctionState(init_sim, -1)
    else: 
        print('Bless you Joel :)')
        
    return init_sim,JS0
```

```python
def setupTrain(training_stimulus,f,onAmp_training=1):

    for i in range(num_drain_training):
        #DRAIN WITHOUT REST
        training_stimulus.append((stimulus__(biasType='Drain',T=run_time,dt=dt)))


    #Sources
    for i in range(num_source_training):
        #SOURCE WITHOUT REST
        training_stimulus.append((stimulus__(biasType='Square',f=f,onAmp=onAmp_training,T=run_time,dt=dt)))
   
    return training_stimulus
```

```python
def setupRest(f,onAmp_rest=0):
    rest_stimulus=[] 
    for i in range(num_drain_training):
        #DRAIN WITHOUT REST
        rest_stimulus.append((stimulus__(biasType='Drain',T=run_time,dt=dt)))


    #Sources
    for i in range(num_source_training):
        #SOURCE WITHOUT REST
        rest_stimulus.append((stimulus__(biasType='DC',f=f,onAmp=onAmp_rest,T=run_time,dt=dt)))
   
    return rest_stimulus
```

```python
#Get the state of the network when all drains are open. 

def getTrainingState(training_stimulus,prevState,drain_pool,sources):
    eles = np.append(drain_pool, sources)
    if len(eles) == num_drain_training + num_source_training:
        training_sim = runSim(connectivity, stimulus = training_stimulus,
                           junctionMode = 'tunneling',
                           dt = dt, T = run_time, 
                           contactMode = 'preSet',
                           electrodes = eles,
                           findFirst = False,
                           start_state = prevState,
                           disable_tqdm=True,
                           criticalFlux=0.01)
        JS1 = getJunctionState(training_sim, -1)
    else: 
        print('Bless you Joel :)')
    return training_sim,JS1
```

```python
def restSim(rest_stimulus,network_state):
    eles = np.append(drain_pool, sources)
    resting_sim = runSim(connectivity, stimulus = rest_stimulus,
                       junctionMode = 'tunneling',
                       dt = dt, T = run_time, 
                       contactMode = 'preSet',
                       electrodes = eles,
                       findFirst = False, 
                       start_state = network_state,disable_tqdm=True)
    resting_state = getJunctionState(resting_sim, -1)

    return resting_state
```

```python
def crossJoel(simulation, sources,
              all_drains): #sim1
    cc = np.zeros(len(all_drains))
    #current flowing through drain / (voltage of sources - voltage of drain)
    for i, d in enumerate(all_drains):
        cc[i]=abs(simulation.electrodeCurrent.T[i][-1]/(simulation.wireVoltage.T[sources[0]][-1]-simulation.wireVoltage.T[d][-1]))
    return cc
```

```python
def change_voltage(testing_stimulus,cross_values, target_index,th,th2):
    n = len(cross_values)
    
    #testing stimulus:
#     onAmp_testing = 1.5 #amplitude for testing
#     signal_time = 0.5
#     run_time = signal_time

    for i in range(n):
        if i in [target_index]: 
            if cross_values[i] < np.abs(th):
                testing_stimulus[i].signal -= 0.005 
        else:
            if cross_values[i] > np.abs(th2):
                testing_stimulus[i].signal += 0.005
    return cross_values,testing_stimulus
```

```python
def runAssLearning(s1_stim,s2_stim,initState):
    simulation,state1=getTrainingState(s2_stim,initState,drain_pool,sources)
    t=crossJoel(simulation,sources,drain_pool)
    cc,s2_stim=change_voltage(s2_stim,t,thisTarget,5e-4,5e-9)
    return cc,s2_stim,state1,simulation
```

```python
#Global Variables
iterations=50
onAmp_init = 1 #amplitude for init
run_time = 2 # time for each iteration
dt = 0.001 #timesteps
numinit = 1 #number of times we initialise network pathways
#Define frequencies for each target
f=[0.5,1,2]

#Initialise Simulations for each target
JS0=[None]*len(target_index); sim1=[None]*len(target_index);sim2=[None]*len(target_index);RS0=[]
sim1_stimulus=[None]*len(target_index);sim2_stimulus=[None]*len(target_index);network_state=[None]*len(target_index)
#Initialise Error Matrix
cross_correlation=np.zeros((len(target_index),iterations,9))


#Method 1) 
#Loop through each target + frequencies to train all:

for currTarget in tqdm(range(len(target_index))):
    thisTarget=target_index[currTarget]
    #NEED TO SAVE STATE AFTER TRAINING ONE FREQUENCY, AND TRAIN WITH NEW FREQUENCY, RATHER THAN START FROM SCRATCH
    #TRYING TO LINK FREQUENCIES WITH PATHWAYS
    
    num_drain_init = len([thisTarget])
    num_source_init = len(sources)

    numtraining = 1 
    num_drain_training = 9 #total number of drain electrodes
    num_source_training = len(sources)
    training_stimulus = []
    if currTarget==0: #if it's the first target/frequency, initialise the network
        sim1_stimulus[currTarget]=init(sources,onAmp_init,f[currTarget])
        sim1[currTarget],JS0[currTarget]=getInitState(sim1_stimulus[currTarget],sources,drain_pool,thisTarget)
    else: #otherwise, take the state from the last network
        JS0[currTarget]=previousTrainedState 
    sim2_stimulus[currTarget]=setupTrain(training_stimulus,f[currTarget],onAmp_init)
    
#     #Perform Training
    for i in tqdm(range(iterations)):
        cross_correlation[currTarget,i,:],sim2_stimulus[currTarget],network_state[currTarget],sim2[currTarget]=runAssLearning(sim1_stimulus[currTarget],sim2_stimulus[currTarget],JS0[currTarget])    
    #add rest after training before saving state for next training
    rest_stimulus=setupRest(f[currTarget])
    RS0.append(restSim(rest_stimulus,network_state[currTarget]))
    previousTrainedState=RS0[currTarget]
```
